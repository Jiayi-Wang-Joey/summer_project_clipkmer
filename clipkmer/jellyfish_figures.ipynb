```python
filename = '/Users/apple/Desktop/Summer_project/dataset/ENCFF170YQV_rep2_2.fq'
kmer_count_6 = jellyfish_count(filename,6)
stats_result_6 = jellyfish_stats()
histo_output_6 = jellyfish_histo()
```

```python
kmer_count_8 = jellyfish_count(filename,8)
stats_result_8 = jellyfish_stats()
histo_output_8 = jellyfish_histo()
```

```python
kmer_count_12 = jellyfish_count(filename,12)
stats_result_12 = jellyfish_stats()
histo_output_12 = jellyfish_histo()
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)

distinct_kmer = [stats_result_6[1], stats_result_8[1], stats_result_12[1]]
ax1.bar(['6-mer','8-mer','12-mer'], distinct_kmer)
ax1.set_ylabel('The number of distinct kmer', fontsize = 12)
ax1.set_title('Distinct kmer',fontsize = 15)
```

```python
fig2 = plt.figure()
ax1 = fig2.add_subplot(1,1,1)

one_count_kmer = [stats_result_6[0], stats_result_8[0], stats_result_12[0]]
ax1.bar(['6-mer','8-mer','12-mer'], one_count_kmer)
ax1.set_ylabel('The number of one-count kmer', fontsize = 12)
ax1.set_title('One-count kmer',fontsize = 15)
```

```python
fig3 = plt.figure()
ax1 = fig3.add_subplot(1,1,1)

max_kmer = [stats_result_6[3], stats_result_8[3], stats_result_12[3]]
ax1.bar(['6-mer','8-mer','12-mer'], max_kmer)
ax1.set_ylabel('The count of max kmer', fontsize = 12)
ax1.set_title('Max kmer',fontsize = 15)
```
