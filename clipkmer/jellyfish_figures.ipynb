```python
import matplotlib.pyplot as plt  
import subprocess
import tempfile
from datetime import datetime
import sys
import os
import csv
from kmer_jellyfish import jellyfish_count, jellyfish_dump
from kmer_freq_positi import kmer_freq_posit, exon_intron
os.chdir('/Users/apple/Desktop/Summer_project/dataset/')
os.environ['PATH'] += ':/Users/apple/bin'
```

```python
filename = 'Homo_sapiens_DNM2_sequence.fa'
length = 8
jf = jellyfish_count('RBFOX2.fa',length)
kmer_count = jellyfish_dump(jf)
freq_posit_dict = kmer_freq_posit(kmer_count,filename,length)
exon_posit = exon_intron('dnm2.gtf')
```

```python
%matplotlib inline
fig1 = plt.figure(figsize = (15,10))
ax1 = fig1.add_subplot(1,1,1)
x_poistion = list(freq_posit_dict[0].keys())
y = list(freq_posit_dict[0].values())
y_count = [ int(x) for x in y ]

ax1.plot(x_poistion,y_count)
ax1.set_xlabel('kmer position',fontsize = 15)
ax1.set_ylabel('count',fontsize = 15)

for exon in exon_posit:
    x = [int(list(exon.keys())[0]),int(list(exon.values())[0])]
    y = [5000,5000]
    ax1.plot(x,y,'r-')

fig1.savefig('frequency.pdf')
```
